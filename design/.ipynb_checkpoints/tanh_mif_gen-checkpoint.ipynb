<font color=0x992f33>**configure**</font>定义了激活函数输入输出的定点小数的格式
- 补码
- 输入输出的整数位宽（最小是1，因为包含了符号位）
- 输入输出的小数位宽

定义了10进制转定点小数的函数`DtoB`和反函数`BtoD`

```python
#################configure##################
#补码
'''Input'''
in_IntWidth   = 3  #Int位宽+符号位
in_FracWidth  = 4 #小数位宽（去尾后的）
'''Output'''
out_IntWidth  = 1 
out_FracWidth = 6
############################################
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pprint




# 定点小数格式：[PureIntWidth].[FracWidth]
in_DataWidth = in_IntWidth + in_FracWidth
in_PureIntWidth = in_IntWidth -1 #不带符号位的Int位宽，即纯Int位宽
out_DataWidth = out_IntWidth + out_FracWidth

pp = pprint.PrettyPrinter(indent=4)

#################function###################
#转成定点二进制补码小数
def DtoB(num,datawidth,fracBits):
    if num >=0:
        num = num*(2**fracBits) #左移，转化为整型
        num = int(num) #舍尾
        e = bin(num)[2:].zfill(datawidth)
    else:
        num = -num
        num = num*(2**fracBits)
        num = round(num)
        if num == 0:
            d = 0
        else :
            d = 2**datawidth - num
        e = bin(d)[2:]
    return e

#定点二进制补码小数转成10进制
def BtoD(bin_num,datawidth,fracBits):
    num = int(bin_num,2)
    if(num)<2**(datawidth-1):
        d = num*2**-fracBits
    else:
        d = (num-2**(datawidth))*2**-fracBits
    return d
```

接下来根据输入输出的定点位宽，计算`tanh`  
图中可以看到准确值和定点之间的误差  

```python
# 计算定点数的范围
max_value = 2**in_PureIntWidth - 2**-in_FracWidth
min_value = -2**in_PureIntWidth

# 生成所有可能的定点数
steps = np.arange(min_value, max_value+2**-in_FracWidth, 2**-in_FracWidth)

##############计算tanh值########################
#理想值
tanh_values = np.tanh(steps)
#量化到定点输出的tanh值
tanh_values_real = [BtoD(DtoB(i,out_DataWidth,out_FracWidth),out_DataWidth,out_FracWidth) for i in tanh_values]

# 绘制结果
plt.figure(figsize=(10, 6))
#plt.scatter(steps, tanh_values, s=5, label="tanh(value)")
plt.plot(steps, tanh_values, label="tanh(ideal value)", linewidth=1)
plt.plot(steps, tanh_values_real, label="tanh(real_value)", linewidth=1)
plt.title(f"Output of tanh for all fixed-point numbers (Format: [{in_IntWidth}].[{in_FracWidth}])")
plt.xlabel("Fixed-point number")
plt.ylabel("tanh(value)")
plt.grid(True)
plt.legend()
plt.show()
```

将结果做成字典，并重新排列，这样方便硬件可以直接索引而不用关心符号问题

```python
result_dict = dict(zip([DtoB(i,in_DataWidth,in_FracWidth) for i in steps],[DtoB(i,out_DataWidth,out_FracWidth) for i in tanh_values_real]))
for key,value in result_dict.items():
    print(key,value)
```

使用`sorted`函数进行重排列  
最终结果存储在<font color=purple font size=5>**sorted_lst**</font>中

```python
sorted_keys = sorted(result_dict.keys())
sorted_lst =[(key,result_dict[key]) for key in sorted_keys]
pp.pprint(sorted_lst)
```

转回十进制小数，检查一下是否为tanh函数  
没有问题

```python
decimal_lst = [(BtoD(i[0],in_DataWidth,in_FracWidth),BtoD(i[1],out_DataWidth,out_FracWidth)) for i in sorted_lst]
plt.figure(figsize=(10, 6))
for point in decimal_lst:
    plt.scatter(point[0], point[1], s=5, label="tanh(value)",c="green")
plt.plot(steps, tanh_values, label="tanh(ideal value)", linewidth=1,c="black")
plt.grid(True)
plt.show()
```
